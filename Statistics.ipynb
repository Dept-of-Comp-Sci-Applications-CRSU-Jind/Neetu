```python
import numpy as np  
  
import matplotlib.pyplot as plt  
x = np.arange(-20,10)  
y = np.power(x,3)  
y1 = np.power(x,3) + np.power(x,2) + x  
plt.scatter(x,y1,c="yellow")  
plt.scatter(x,y)  
```
